```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("loan.csv")
df
```

```python
df.drop(columns=['Loan_ID'], inplace=True)
df.info()
```

```python
df.isnull().sum()
```

```python
categorical_cols=df.select_dtypes(include=['object']).columns
imputer_cat = SimpleImputer(strategy='most_frequent')
df[categorical_cols]= imputer_cat.fit_transform(df[categorical_cols])
```

```python
df.head()
```

```python
numerical_cols=df.select_dtypes(include='number').columns
imputer_num=SimpleImputer(strategy='median')
df[numerical_cols]=imputer_num.fit_transform(df[numerical_cols])
```

```python
df.head()
```

```python
label_encoders = {}
for col in categorical_cols:
    label_encoders[col]=LabelEncoder()
    df[col]=label_encoders[col].fit_transform(df[col])
```

```python
df.isnull().sum()
```

```python
df['Loan_Status'].value_counts()
```

```python
df['Loan_Status'].value_counts(normalize=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['Loan_Status']) #Features
y = df['Loan_Status'] #target variable
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='liblinear')
rfe_selector=RFE(estimator=log_reg,n_features_to_select=5)
X_rfe=rfe_selector.fit_transform(X,y)
rfe_selected_features=X.columns[rfe_selector.get_support()]
print("RFE selected Features:",rfe_selected_features.tolist())
```

```python
X_rfe
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_rfe, y , test_size=0.2, stratify=y, random_state=42)
```

```python
print("X_train:",X_train.shape)
print("X_test:",X_test.shape)
print("y_train:",y_train.shape)
print("y_train:",y_test.shape)
```

```python
nb_model=GaussianNB()
nb_model.fit(X_train,y_train)
```

```python
y_pred=nb_model.predict(X_test)
y_pred
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Classification Report:\n",classification_report(y_test,y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test, y_pred))
```
